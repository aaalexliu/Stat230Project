```python
import json
import pprint
pp = pprint.PrettyPrinter(indent=4)
```

```python
with open('full_data_trimmed.json') as fp:
    full = json.load(fp)
```

```python
pp.pprint(full[198]['full'])
```

```python
for meal in full:
    if meal['type'] != 'Breakfast':
        meal['traditional'] = meal['full'].get('Traditional')
    else:
        meal['traditional'] = meal['full'].get('Breakfast/Grill')
```

```python
import csv
keys = ['date', 'type', 'traditional']
```

```python
import datetime
with open('val_traditional_{time}.csv'.format(time = datetime.datetime.now().strftime('%H-%M-%S')),'w+') as output_file:
    dict_writer = csv.DictWriter(output_file, keys, extrasaction='ignore')
    dict_writer.writeheader()
    dict_writer.writerows(full)
```
